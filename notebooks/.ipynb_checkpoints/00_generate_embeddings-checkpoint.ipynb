```python
import pandas as pd

import os

from dotenv import load_dotenv
from openai import OpenAI
from tqdm import tqdm
```

```python
# Load API key from .env file
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key = OPENAI_API_KEY)
if not OPENAI_API_KEY:
    raise ValueError("OpenAI API key not found! Make sure it's in the .env file.")

EMBEDDING_MODEL = "text-embedding-3-small"
GPT_MODEL = "gpt-3.5-turbo"    
```

```python
def get_embedding(text, model= EMBEDDING_MODEL):
    return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
# Loading the dataset
df = pd.read_csv("/home/lizette/Documents/udacity_LLMs/project/books_data.csv")
```

```python
df.head()
```

```python
df = df.sample(n = 10)

#Drop nans
df = df.dropna(subset = ["description", "categories"])

df["embedding"] = df["description"].apply(lambda x: get_embedding(x, model='text-embedding-3-small'))

df = df.rename(columns={"description": "text"})
df.to_csv("books_data_sample_with_embeddings.csv")
```
